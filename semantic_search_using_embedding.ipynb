```python
search_term = 'cat'
search_term_vec = get_embedding(search_term,engine='text-embedding-3-small')
print(len(search_term_vec))
```

```python
from openai.embeddings_utils import cosine_similarity

df['simalarities'] = df['embedding'].apply(lambda x : cosine_similarity(x, search_term_vec))
df.head()
```

```python
df.sort_values("simalarities",ascending=False)
```
