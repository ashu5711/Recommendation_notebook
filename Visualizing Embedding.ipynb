```python
import openai
import pandas as pd
import numpy as np
from nomic import atlas
```

```python
embedding_cache_file = './books_embedding.csv'
df = pd.read_csv(embedding_cache_file)
df['embedding'] = df['embedding'].apply(eval).apply(np.array)
```

```python
data = df[['title', 'authors', 'categories']].to_dict('records')
data
```

```python
embeddings = list(df['embedding'])
embeddings = np.array(embeddings)

project = atlas.map_data(
    embeddings=embeddings,
    data=data,
    is_public=True,identifier="books",description="books title cluster"

)
```
